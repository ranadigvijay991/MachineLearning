```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features=1, n_informative=1,n_redundant=0, n_repeated=0,n_classes=2,n_clusters_per_class=1)
print(X)
print(y)
```

```python
print(X.shape,y.shape)
```

```python
X = np.reshape(X,(1,100))
y = np.reshape(y,(100,1))
```

```python
def sigmoid(y_pred):
    s = 1/(1+np.exp(-y_pred))
    return s
```

```python
 def descent(X,y):
    m = 100
    w = np.zeros((1,m))
    b = 0 
    lr = 0.01
    for i in range(m):
        z = np.dot(X,w.T)+b
        a = sigmoid(z)
        cost = (-1/m)*(np.sum((y*np.log(a)) + ((1-y)*(np.log(1-a)))))
        dw = (1/m)*(np.dot(X,(a-y)))
        db = (1/m)*np.sum(a-y)
        
        w = w - (lr * dw)
        b = b - (lr * db)
        print("Iteration : ",i,"Cost : ",cost,"w : ",w[0][i],"b : ",b)
```

```python
sigmoid(1)
```

```python
descent(X,y)
```
